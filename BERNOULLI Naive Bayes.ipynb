```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
reviews=[
    'this movie was great and fantastic', #positive
    'absolutely wonderful film',
    "i loved the performance and story",
    "terrible movie,very boring",
    "waste of time,do not watch" ,#neg
    "poorly made and bad acting"
]
```

```python
labels=["positive","positive","positive","negative","negative","negative"]
```

```python
vectorizer= CountVectorizer(binary=True)
x=vectorizer.fit_transform(reviews)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,labels,test_size=0.33,random_state=42)
```

```python
model=BernoulliNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print("accuracy:",accuracy_score(y_test,y_pred))
```

```python
new_reviews=["Fantastic acting and great strory","very boring and badly done"]
x_new=vectorizer.transform(new_reviews)
predictions=model.predict(x_new)
```

```python
for review,pred in zip(new_reviews,predictions):
    print(f"Review: \"{review}\"  --> sentiment:{pred}")
```
