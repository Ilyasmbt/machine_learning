```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,_= make_blobs(n_samples=300,centers=4,cluster_std=0.60,random_state=42)
```

```python
Kmeans=KMeans(n_clusters=4,random_state=42)
Kmeans.fit(x)
```

```python
new_data=np.array([[2,1],[4,4]])
predicted_cluster =Kmeans.predict(new_data)
print("Predicted cluster labels for new data:",predicted_cluster)
```

```python
plt.scatter(x[:,0],x[:,1],c=Kmeans.labels_,cmap="viridis",alpha=0.6)
plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,1],s=200,c="red",marker="X",label="Centroids")
plt.scatter(new_data[:,0],new_data[:,1],c="black",marker="*",s=200,label="New Data")
plt.legend()
plt.title("K-means prediction")
plt.show()
```
